```python
# By Waleed Alsanie 2021. Plotting the negative of a mixture of gaussians of four components. 
# Adapted from: 
# https://notebook.community/empet/Plotly-plots/Gaussian-Mixture
# To plot a hypothetical objective function
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import multivariate_normal as bn
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm
# For the arrows:
from matplotlib.patches import FancyArrowPatch
from mpl_toolkits.mplot3d.proj3d import proj_transform

######################################################################
# To draw 3d Arrows. taken  from:
# https://gist.github.com/WetHat/1d6cd0f7309535311a539b42cccca89c
class Arrow3D(FancyArrowPatch):
    def __init__(self, x, y, z, dx, dy, dz, *args, **kwargs):
        super().__init__((0,0), (0,0), *args, **kwargs)
        self._xyz = (x,y,z)
        self._dxdydz = (dx,dy,dz)

    def draw(self, renderer):
        x1,y1,z1 = self._xyz
        dx,dy,dz = self._dxdydz
        x2,y2,z2 = (x1+dx,y1+dy,z1+dz)

        xs, ys, zs = proj_transform((x1,x2),(y1,y2),(z1,z2), renderer.M)
        self.set_positions((xs[0],ys[0]),(xs[1],ys[1]))
        super().draw(renderer)
        
def _arrow3D(ax, x, y, z, dx, dy, dz, *args, **kwargs):
    '''Add a 3d arrow to an `Axes3D` instance.'''

    arrow = Arrow3D(x, y, z, dx, dy, dz, *args, **kwargs)
    ax.add_artist(arrow)

setattr(Axes3D,'arrow3D',_arrow3D)

######################################################################
# Construct the covariance matrix
def cov(sigma, corr_cof) -> np:
    covar = corr_cof * sigma[0] * sigma[1]
    return np.array([[sigma[0] ** 2, covar], [covar, sigma[1] ** 2]])

# A returns a mixture of bivariate gaussian
def mixture_gaussian(x, y, w, means, cov_mat) -> np:
    # Add another dimension of two components for bi random vector. 
    bi_random = np.empty(x.shape + (2,))
    bi_random[:, :, 0] = x; 
    bi_random[:, :, 1] = y 
    z = np.zeros(x.shape)
    for k in range(4) :
        z = z + w[k] * bn.pdf(bi_random, mean=means[k,:], cov=cov_mat[:,:, k])
    return z

#####################################################################
# Weights associated with the mixture components
w = [0.3, 0.2, 0.2, 0.3]

#####################################################################
# Means
means = np.array([[10, 10], [20, 5], [25, 22],[4, 15]])
# Standard deviations
sigmas = np.array([[4, 4],   [5, 3],  [3, 5],[3, 3.5]]) 
# Correlation coefficients
rho = np.array([0.0, -0.67,-0.76,0.5])
# Initialise the covariance matrix
cov_matrix=np.zeros((2,2,4), float)

# Fill the covariance matrix
for k in range(4):
    cov_matrix[:,:,k] = cov(sigmas[k,:], rho[k])

####################################################################
# Plotting
fig, ax = plt.subplots(subplot_kw={"projection": "3d"})
fig.set_size_inches(10, 10)
ax.set_xticks([])      # Shut the x-axis numbers
ax.set_yticks([])      # Shut the y-axis numbers
ax.set_zticks([])      # Shut the z-axis numbers

xx = np.linspace(0, 30, 100)
yy = np.linspace(0, 30, 100)
x,y = np.meshgrid(xx, yy)

# Inverse the mixture
z = -mixture_gaussian(x, y, w,  means, cov_matrix)

ax.set_title('Hypothetical Surface of an Objective Function')
ax.plot_surface(x, y, z, rstride=1, cstride=1, cmap=cm.RdBu, linewidth=0, antialiased=False, zorder=1)
####################################################################
# Arrows:
#ax.arrow3D(1,4.5,-0.0012, 3,14,-0.0003, mutation_scale=20, arrowstyle="-|>", 
#           ec ='red', fc='red', lw=4,zorder=3)
#ax.arrow3D(2,4.5,-0.0013, 1,1,-0.0001, mutation_scale=10, arrowstyle="-|>", 
#           ec ='black', fc='black', lw=3,zorder=3)
#
####################################################################
# Saddlepoint
#ax.scatter(5, 5, -0.003, color='black', marker='o', zorder=50)
##################################################################
# Set the view
ax.view_init(28, -21)
##################################################################
# Save and Show 
#plt.savefig('gradient_descent.eps')
#plt.savefig('gradient_descent_no_arrows.eps')
#plt.savefig('vanishing_exploding_gradients.eps')
plt.show()
```
